```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.DataFrame(pd.read_csv("WA_Marketing-Campaign.csv"))
```

```python
data
```

## Turning different Marketsize into different dataframes
Doing this is combat the inherent differences of larger markets compared to smaller locations

```python
data_small = data[data["MarketSize"] == "Small"]
data_medium = data[data["MarketSize"] == "Medium"]
data_large = data[data["MarketSize"] == "Large"]
```

```python
def boxplots(data, marketsizename="MarketSize"):
    plt.boxplot(x=[data[data["Promotion"] == p]["SalesInThousands"] for p in data["Promotion"].unique()], labels=data["Promotion"].unique())
    plt.title(marketsizename)
    plt.xlabel("Promotion")
    plt.ylabel("SalesInThousands")
    plt.show()

boxplots(data_small, "Small")
boxplots(data_medium, "Medium")
boxplots(data_large, "Large")
```
